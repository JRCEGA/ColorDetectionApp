### <span style="color:blue">Problema de Investigación</span>

El problema de investigación se centra en la capacidad de analizar la coloración de las hojas de una planta como indicador de su estado de salud y, en la posibilidad de determinar si la planta necesita agua, se encuentra en buen estado, o tiene agua en exceso. Este problema es de suma importancia para la gestión eficiente de la agricultura y la jardinería, y para abordar esta problemática, se propone utilizar un hardware específico que incluye el sensor de color EZO-RGB, un protocolo de datos UART, y un microcontrolador Arduino UNO R3.

Algunos aspectos clave para la investigación son:

Importancia de la Salud de las Plantas.
Indicadores Visuales.
Tecnología de Sensores.
Aplicaciones Prácticas.
Sostenibilidad Agrícola.

*Descripción*

El Pothos, también conocido como los nombres antes mencionados, es una planta de hoja Perenne que tiene hojas verdes con manchas amarillas. Las hojas son ovaladas y pueden crecer hasta 12 pulgadas de largo. La planta puede crecer hasta 30 pies de largo en su hábitat natural.

El Pothos es una planta muy resistente que puede crecer en una variedad de condiciones. La planta prefiere la luz indirecta, pero también puede tolerar la luz solar directa durante períodos cortos de tiempo. El Pothos se debe regar cuando la tierra esté seca al tacto. La planta no requiere una fertilización regular.

Para saber si el Pothos está seco o le falta agua, puede observar su color. Las hojas de un Pothos sano deben ser de color verde brillante. Si las hojas comienzan a ponerse amarillas o marrones, es una señal de que la planta está deshidratada.

Se podría determinar el estado del Pothos en base a su color:


*   Hojas verdes brillantes: está sano y bien hidratado.
*   Hojas amarillas o marrones: está deshidratado.
*   Hojas marchitas: está gravemente deshidratado y necesita agua urgentemente.
*   Hojas con manchas: puede estar sufriendo de una enfermedad o plaga.

### <span style="color:blue">Parámetros en ArduinoIDE</span>

1. Puertos RX y TX: se utilizan los pines 2 y 3 como pines RX y TX para la comunicación serial.
2. SoftwareSerial: se utiliza la biblioteca SoftwareSerial para habilitar la comunicación serial adicional en los pines definidos.
3. Inicializar SoftwareSerial: se inicializa un objeto llamado "myserial" de la clase SoftwareSerial con los pines RX y TX.
4. Baud Rate: se establece en 9600 para las comunicaciones seriales (H y S).
5. Buffers de entrada: se reservan los bytes de memoria para almacenar cadenas de entrada desde la laptop y desde el dispositivo (sensor).
6. serialEvent: se ejecutará cuando el puerto serial del hardware reciba un carácter.
7. loop: verifica si se completa la cadena de entrada (carácteres) desde la laptop.
8. print_RGB_data: Analiza la cadena recibida desde el sensor y lo imprime en el monitor.
9. Puerto 0 y 1: se utilizan los pines 0 y 1 como pines RX y TX para la comunicación serial del hc-05.

> Definición de puertos

* RX = 3 Cable blanco mostrado en prototipo. Este pin se utiliza para recibir datos. En un sensor de color, puede usarse para recibir información de color o datos relacionados con la detección de color.
* TX = 2 Cable verde mostrado en prototipo. Este pin se utiliza para transmitir datos. Puede ser utilizado para enviar datos desde el sensor de color a otro dispositivo, como un microcontrolador o una computadora.
* GND = Cable negro mostrado en prototipo. Este pin se conecta a la tierra (ground) y sirve como referencia de voltaje cero para el circuito. Es esencial conectar este pin a la tierra del sistema para establecer un nivel común de referencia.
* VCC = 5V cable rojo mostrado en prototipo. Este pin se refiere a la alimentación positiva. Se coneca a una fuente de voltaje para suministrar energía al sensor. La tensión exacta requerida dependerá de las especificaciones del sensor, pero comúnmente se utiliza 3.3V o 5V.

* Data bit: 8-bit

* Baud rate: 9600 bits/s

* Parity: None

* Step bit: 1 stop bit

* 12 caracteres de ingreso

> Voltaje

El voltaje que necesita el EZO-RGB está dentro del rango 3.3V-5V, por lo que se plantea trabajar con un voltaje de 5V al conectar el sensor al arduino, y el voltaje del arduino a la laptop es de 3.3V.

¿Con qué periocidad se lee? La velocidad del reloj va en 100 kHZ.

#### Configuración del sensor EZO-RGB con Arduino UNO R3.



```
#include <SoftwareSerial.h>                           // debemos incluir la biblioteca SoftwareSerial
#define rx 2                                          // definir el pin rx
#define tx 3                                          // definir el pin tx

SoftwareSerial myserial(rx, tx);                      // definir cómo funcionará el puerto serie virtual

String inputstring = "";                              // una cadena para almacenar datos entrantes desde la PC
String sensorstring = "";                             // una cadena para almacenar datos del producto Atlas Scientific
boolean input_string_complete = false;                // ¿hemos recibido todos los datos de la PC?
boolean sensor_string_complete = false;               // ¿hemos recibido todos los datos del producto Atlas Scientific?

void setup() {                                        // configurar el hardware
  Serial.begin(9600);                                 // establecer la velocidad de baudios para el puerto serie hardware a 9600
  myserial.begin(9600);                               // establecer la velocidad de baudios para el puerto serie virtual a 9600
  inputstring.reserve(10);                            // reservar algunos bytes para recibir datos de la PC
  sensorstring.reserve(30);                           // reservar algunos bytes para recibir datos del producto Atlas Scientific
}

void serialEvent() {                                  // si el puerto serie hardware recibe un carácter
  inputstring = Serial.readStringUntil(13);           // leer la cadena hasta que veamos un <CR>
  input_string_complete = true;                       // establecer la bandera que indica si hemos recibido una cadena completa de la PC
}

void loop() {                                        
  if (input_string_complete == true) {                // si se ha recibido una cadena completa de la PC
    myserial.print(inputstring);                      // enviar esa cadena al producto Atlas Scientific
    myserial.print('\r');                             // agregar un <CR> al final de la cadena
    inputstring = "";                                 // borrar la cadena
    input_string_complete = false;                    // restablecer la bandera que indica si hemos recibido una cadena completa de la PC
  }

  if (myserial.available() > 0) {                     // si vemos que el producto Atlas Scientific ha enviado un carácter
    char inchar = (char)myserial.read();              // obtener el carácter que acabamos de recibir
    sensorstring += inchar;                           // agregar el carácter a la variable llamada sensorstring
    if (inchar == '\r') {                             // si el carácter entrante es un <CR>
      sensor_string_complete = true;                  // establecer la bandera
    }
  }

  if (sensor_string_complete == true) {               // si se ha recibido una cadena completa del producto Atlas Scientific
    if (isdigit(sensorstring[0]) == false) {          // si el primer carácter en la cadena no es un dígito
      Serial.println(sensorstring);                   // enviar esa cadena al monitor serial de la PC
    }
    else                                              // si el primer carácter en la cadena NO es un dígito
    {
      print_RGB_data();                               // entonces llamar a esta función
    }
    sensorstring = "";                                // borrar la cadena
    sensor_string_complete = false;                   // restablecer la bandera que indica si hemos recibido una cadena completa del producto Atlas Scientific
  }
}

void print_RGB_data(void) {                           // esta función analizará la cadena

  char sensorstring_array[30];                        // creamos un array de caracteres
  char *red;                                          // puntero de caracteres utilizado en el análisis de cadenas
  char *grn;                                          // puntero de caracteres utilizado en el análisis de cadenas
  char *blu;                                          // puntero de caracteres utilizado en el análisis de cadenas
  int int_red;                                        // utilizado para almacenar un entero que es el color rojo
  int int_grn;                                        // utilizado para almacenar un entero que es el color verde
  int int_blu;                                        // utilizado para almacenar un entero que es el color azul
  
  sensorstring.toCharArray(sensorstring_array, 30);   // convertir la cadena en un array de caracteres
  red = strtok(sensorstring_array, ",");              // analicemos el array en cada coma
  grn = strtok(NULL, ",");                            // analicemos el array en cada coma
  blu = strtok(NULL, ",");                            // analicemos el array en cada coma
  
  Serial.print(red);                                  // imprimir el valor rojo
  Serial.print(",");                                  // imprimir el separador de coma

  Serial.print(grn);                                  // imprimir el valor verde
  Serial.print(",");                                  // imprimir el separador de coma

  Serial.print(blu);                                  // imprimir el valor azul y pasar a la línea siguiente
   
// int_red= atoi(red);                                 // descomentar esta línea para convertir el carácter en un entero
// int_grn= atoi(grn);                                 // descomentar esta línea para convertir el carácter en un entero
// int_blu= atoi(blu);                                 // descomentar esta línea para convertir el carácter en un entero
}

```

```
var bluetoothApp = {
        macAddress: "98:D3:02:96:C3:41", // Asegúrate de que esta es la dirección MAC correcta
        isConnected: false,

        setupConnectButton: function() {
            var connectButton = document.getElementById('connectButton');
            if (connectButton) {
                connectButton.addEventListener('click', function() {
                    if (bluetoothApp.isConnected) {
                        bluetoothApp.disconnectFromBluetoothDevice();
                    } else {
                        bluetoothApp.connectToBluetoothDevice();
                    }
                });
            }
        },

        connectToBluetoothDevice: function() {
            var permissions = cordova.plugins.permissions;
            permissions.hasPermission(permissions.BLUETOOTH_CONNECT, function(status) {
                if (!status.hasPermission) {
                    permissions.requestPermission(permissions.BLUETOOTH_CONNECT, function(status) {
                        if (status.hasPermission) {
                            // Ahora comprueba el permiso BLUETOOTH_SCAN
                            permissions.hasPermission(permissions.BLUETOOTH_SCAN, function(status) {
                                if (!status.hasPermission) {
                                    permissions.requestPermission(permissions.BLUETOOTH_SCAN, function(status) {
                                        if (status.hasPermission) {
                                            bluetoothSerial.connect(bluetoothApp.macAddress, bluetoothApp.onConnectSuccess, bluetoothApp.onConnectError);
                                        }
                                    }, function() {
                                        console.warn('BLUETOOTH_SCAN permission denied');
                                    });
                                } else {
                                    bluetoothSerial.connect(bluetoothApp.macAddress, bluetoothApp.onConnectSuccess, bluetoothApp.onConnectError);
                                }
                            });
                        }
                    }, function() {
                        console.warn('BLUETOOTH_CONNECT permission denied');
                    });
                } else {
                    // Si ya tiene el permiso BLUETOOTH_CONNECT, comprueba BLUETOOTH_SCAN
                    permissions.hasPermission(permissions.BLUETOOTH_SCAN, function(status) {
                        if (!status.hasPermission) {
                            permissions.requestPermission(permissions.BLUETOOTH_SCAN, function(status) {
                                if (status.hasPermission) {
                                    bluetoothSerial.connect(bluetoothApp.macAddress, bluetoothApp.onConnectSuccess, bluetoothApp.onConnectError);
                                }
                            }, function() {
                                console.warn('BLUETOOTH_SCAN permission denied');
                            });
                        } else {
                            bluetoothSerial.connect(bluetoothApp.macAddress, bluetoothApp.onConnectSuccess, bluetoothApp.onConnectError);
                        }
                    });
                }
            });
        },

        disconnectFromBluetoothDevice: function() {
            bluetoothSerial.disconnect(bluetoothApp.onDisconnectSuccess, bluetoothApp.onDisconnectError);
        },

        onConnectSuccess: function() {
            bluetoothApp.isConnected = true;
            bluetoothApp.updateConnectButton();
            document.getElementById('statusMessage').textContent = "Conectado!";
            localStorage.setItem('isBluetoothConnected', 'true');
            updateBluetoothStatus(true);
            // Aquí puedes continuar con otras operaciones después de la conexión
        },

        onConnectError: function(error) {
            bluetoothApp.isConnected = false;
            bluetoothApp.updateConnectButton();
            document.getElementById('statusMessage').textContent = "Error de conexión: " + error;
        },

        onDisconnectSuccess: function() {
            bluetoothApp.isConnected = false;
            bluetoothApp.updateConnectButton();
            document.getElementById('statusMessage').textContent = "Desconectado!";
            localStorage.setItem('isBluetoothConnected', 'false');
            updateBluetoothStatus(false);
        },

        onDisconnectError: function(error) {
            document.getElementById('statusMessage').textContent = "Error al desconectar: " + error;
        },

        updateConnectButton: function() {
            var connectButton = document.getElementById('connectButton');
            if (bluetoothApp.isConnected) {
                connectButton.textContent = "Desconectar Bluetooth";
            } else {
                connectButton.textContent = "Conectar Bluetooth";
            }
        }
        
    };

    bluetoothApp.setupConnectButton();
```
